## Section 5 continued: Exploratory Data Analysis (EDA)
Before feeding the images into our YOLOv8 model, it is crucial to understand the distribution of our dataset and visualize some samples.

In this section, we will:
1. **Analyze the Dataset Split:** Check how many images are allocated for Training vs. Validation.
2. **Visualize Sample Images:** Display a grid of random road images from our dataset to understand the environmental conditions (lighting, road types, pothole shapes).

```python
import os
import glob
import cv2
import matplotlib.pyplot as plt

# 1. Define folder paths for training and validation images
train_images_path = "pothole_dataset_yolo/train/images"
valid_images_path = "pothole_dataset_yolo/valid/images"

# 2. Count the total number of images in each folder
train_count = len(os.listdir(train_images_path))
valid_count = len(os.listdir(valid_images_path))

print(f"Total Training Images: {train_count}")
print(f"Total Validation Images: {valid_count}\n")

# 3. Create a Bar Chart to show the dataset split
plt.figure(figsize=(8, 5))
bars = plt.bar(['Training Data', 'Validation Data'], [train_count, valid_count], color=['#2ca02c', '#1f77b4'])
plt.title('Dataset Distribution (Train vs Validation)', fontsize=14, fontweight='bold')
plt.ylabel('Number of Images', fontsize=12)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 20, int(yval), ha='center', fontsize=12, fontweight='bold')

plt.show()
print("-" * 60)

# 4. Show a grid of 6 sample images from the training dataset
print("\n📷 Visualizing Sample Images from the Dataset:")
image_files = glob.glob(os.path.join(train_images_path, "*.*"))

plt.figure(figsize=(15, 8))
for i in range(min(6, len(image_files))):
    # Read image and convert color format from BGR (OpenCV default) to RGB
    img = cv2.imread(image_files[i])
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Plot in a 2x3 grid
    plt.subplot(2, 3, i + 1)
    plt.imshow(img)
    plt.axis('off')
    plt.title(f"Sample Image {i+1}", fontsize=12)

plt.tight_layout()
plt.show()
```

## Section 6: Model Training & Evaluation
In this section, we train the YOLOv8 (Nano) model on our custom pothole dataset.
* We dynamically generate the `data.yaml` configuration file to ensure accurate absolute paths.
* **Epochs:** 20 (Sufficient for learning basic pothole features without overfitting).
* **Image Size:** 640x640 pixels (YOLO standard).

```python
import yaml
import os
from ultralytics import YOLO

dataset_path = '/content/pothole_dataset_yolo'

data_yaml = {
    'train': os.path.join(dataset_path, 'train/images'),
    'val': os.path.join(dataset_path, 'valid/images'),
    'nc': 1,
    'names': ['pothole']
}

yaml_file_path = os.path.join(dataset_path, 'data.yaml')
with open(yaml_file_path, 'w') as f:
    yaml.dump(data_yaml, f, default_flow_style=False)

print("data.yaml file created successfully!")
print("Starting YOLOv8 Model Training. This will take a few minutes...\n")

model = YOLO('yolov8n.pt')

results = model.train(
    data=yaml_file_path,
    epochs=20,                 # Train for 20 rounds
    imgsz=640,                 # Standard image size
    batch=16,                  # Process 16 images at a time
    project='Pothole_Project',
    name='YOLOv8_Training'
)

print("\nTraining Complete The AI has learned to detect potholes.")
```

### Visualizing Training Metrics & Running Inference
As per the requirements, we will now:
1. Display the **Training Curves** (Loss and mAP) to evaluate model learning.
2. Run **Inference** on a random validation image to see the model detect potholes in real-time.

```python
import os
import glob
import random
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO

print("Starting Section 7 (Segmentation) & Section 8 (Classification)...\n")

# 1. Pick a random validation image from our pothole dataset
valid_images = glob.glob('pothole_dataset_yolo/valid/images/*.jpg')
test_image = random.choice(valid_images)
print(f"Testing on image: {os.path.basename(test_image)}")

print("\n1. Running Segmentation (Section 7)...")
seg_model = YOLO('yolov8n-seg.pt')
seg_results = seg_model.predict(source=test_image, save=True, project='Pothole_Project', name='Seg_Inference', exist_ok=True)

# Display the Segmentation result image
seg_img_path = os.path.join(seg_results[0].save_dir, os.path.basename(test_image))
if os.path.exists(seg_img_path):
    img_seg = cv2.imread(seg_img_path)
    img_seg = cv2.cvtColor(img_seg, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(10, 6))
    plt.imshow(img_seg)
    plt.title('Section 7: Segmentation Results (Masks)', fontsize=14, fontweight='bold')
    plt.axis('off')
    plt.show()

print("-" * 60)
print("\n2. Running Classification Top-5 (Section 8)...")
cls_model = YOLO('yolov8n-cls.pt')
cls_results = cls_model.predict(source=test_image, save=True, project='Pothole_Project', name='Cls_Inference', exist_ok=True)

# Extract and neatly print the Top 5 Class predictions
print("\nTop-5 Classification Results (ImageNet Categories):")
probs = cls_results[0].probs
top5_indices = probs.top5
top5_conf = probs.top5conf

for i in range(5):
    class_name = cls_results[0].names[top5_indices[i]]
    confidence = top5_conf[i].item() * 100
    print(f"{i+1}. {class_name.capitalize()}: {confidence:.2f}%")
```

## Section 9 & 10: Variant Benchmarking & Hyperparameter Experiments

To ensure optimal performance for our Smart City Application, we conduct the following experiments:
1. **Variant Benchmarking:** Comparing inference speeds of YOLOv8n (Nano) vs YOLOv8s (Small) to find the best balance between speed and accuracy.
2. **Hyperparameter Tuning:** Running two short experiments with different Optimizers (`AdamW` vs `SGD`) and Learning Rates to observe the impact on initial training loss.

```python
import time
import os
import pandas as pd
from ultralytics import YOLO

print("Starting Section 9: Variant Benchmark Comparison...\n")

model_nano = YOLO('yolov8n.pt')
model_small = YOLO('yolov8s.pt')
test_img = 'pothole_dataset_yolo/valid/images/' + os.listdir('pothole_dataset_yolo/valid/images')[0]

start_time = time.time()
model_nano.predict(test_img, verbose=False)
nano_time = (time.time() - start_time) * 1000

start_time = time.time()
model_small.predict(test_img, verbose=False)
small_time = (time.time() - start_time) * 1000

data = {'Model Variant': ['YOLOv8 Nano (n)', 'YOLOv8 Small (s)'],
        'Inference Time (ms)': [f"{nano_time:.2f}", f"{small_time:.2f}"],
        'Suitability': ['Best for Edge Devices / Mobile', 'Better for Cloud Servers']}
df_benchmark = pd.DataFrame(data)
print("Benchmark Results:")
print(df_benchmark.to_markdown(index=False))
print("-" * 60)

print("\nStarting Section 10: Hyperparameter Experiments (1 Epoch each)...")
yaml_file_path = 'pothole_dataset_yolo/data.yaml'

# 5. Experiment 1: AdamW Optimizer with High Learning Rate
print("\n▶Running Experiment 1 (Optimizer: AdamW, LR: 0.01)...")
exp1 = YOLO('yolov8n.pt').train(
    data=yaml_file_path, epochs=1, imgsz=640, batch=16,
    optimizer='AdamW', lr0=0.01, project='Pothole_Project', name='Exp1_AdamW'
)

# 6. Experiment 2: SGD Optimizer with Lower Learning Rate
print("\n▶Running Experiment 2 (Optimizer: SGD, LR: 0.001)...")
exp2 = YOLO('yolov8n.pt').train(
    data=yaml_file_path, epochs=1, imgsz=640, batch=16,
    optimizer='SGD', lr0=0.001, project='Pothole_Project', name='Exp2_SGD'
)

print("\nSections 9 & 10 Completed Successfully Experiments saved.")
```

## Section 11: Model Export for Deployment (ONNX)
To make our pothole detection model ready for real-world deployment (e.g., on edge devices, mobile apps, or cloud servers), we will export the trained PyTorch weights (`.pt`) to the highly interoperable **ONNX** (Open Neural Network Exchange) format.

```python
import os
import glob
from ultralytics import YOLO

print("Searching for your trained model (best.pt)...")
model_files = glob.glob('**/best.pt', recursive=True)

if model_files:
    target_model_path = model_files[0]
    print(f"Found model at: {target_model_path}")
    model = YOLO(target_model_path)

    # Export to ONNX
    print("Exporting to ONNX format... Please wait.")
    model.export(format='onnx', dynamic=True, opset=12)

    print(f"\nSUCCESS! model is now exported to ONNX.")
else:
    print("ERROR: Could not find 'best.pt'.")
    print("Please make sure Section 6 (Training) was completed successfully.")
```
